### Get weekly points

```python
def create_weekj_df(week):
    # read csv
    filename = f"Raw Data/Game Weeks/gw{week}.csv"
    weekj_df = pd.read_csv(filename, index_col="name")
    
    # delete rows with repeated indexes and keep first occurence
    weekj_df = weekj_df[~weekj_df.index.duplicated(keep='first')]

    # extract whether he played and total points for the week
    weekj_df = weekj_df[["minutes", "total_points"]]
    weekj_df ["minutes"] = np.where(weekj_df["minutes"]>0, 1, 0)
    weekj_df = weekj_df.rename(columns={"total_points": f"total_points week {week}","minutes": f"played week {week}"})
    
    # returned points_j_df and played_j_df
    points_j_df = weekj_df[[f"total_points week {week}"]]
    played_j_df = weekj_df[[f"played week {week}"]]
    return points_j_df, played_j_df
create_weekj_df(1)[1]
```

```python
points_df, played_df = create_weekj_df(1)
# loop through all weeks
for j in range(2,38+1):
    points_j_df, played_j_df = create_weekj_df(j)
    points_df = pd.merge(points_df, points_j_df, left_index=True, right_index=True, how="outer")
    played_df = pd.merge(played_df, played_j_df, left_index=True, right_index=True, how="outer")

points_df = points_df.fillna(0)
played_df = played_df.fillna(0)
points_df
#played_df
```

```python
week_df = pd.merge(points_df, played_df, left_index=True, right_index=True, how="outer")
week_df
```

```python
# merge with main df
df = pd.merge(df, week_df, left_index=True, right_index=True, how="outer")
df
```

```python
df.to_csv("data.csv")
```
