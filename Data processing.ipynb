```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Raw Data/cleaned_players.csv")
df.head()
```

```python
df["full name"] = df["first_name"] + " " + df["second_name"]
df.head()
```

```python
df = df[["full name", "total_points", "element_type", "now_cost"]]
df.head()
```

```python
df["GK"] = np.where(df.element_type == "GK",1, 0)
df["DEF"] = np.where(df.element_type == "DEF",1, 0)
df["MID"] = np.where(df.element_type == "MID",1, 0)
df["FWD"] = np.where(df.element_type == "FWD",1, 0)

df
```

```python
df = df.set_index("full name")
df.head()
```

### Add club of players

```python
club_df = pd.read_csv("Raw Data/gw1.csv", index_col="name")
club_df.head()
```

```python
club_df = club_df[["team"]]
club_df.head()
```

```python
output_df = df.combine_first(club_df)
output_df
```

```python
output_df = output_df.dropna()
output_df.head()
```

```python
output_df.to_csv("data.csv")
```
